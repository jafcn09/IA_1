```python
import os.path as osp

import mmcv
import numpy as np

#from mmdet.datasets.base_det_dataset import DATASETS
from mmdet.registry import DATASETS
#from mmdet.registry import CustomDataset
#from mmdet.base_det_dataset import BaseDetDataset
from mmdet.datasets.base_det_dataset import BaseDetDataset



@DATASETS.register_module()
class KittiTinyDataset2(BaseDetDataset):
    CLASSES = ('cataract_eye', 'non_cataract_eye')

    def __init__(self, ann_file, img_prefix):
        self.ann_file = ann_file
        self.img_prefix = img_prefix

    def load_annotations(self):
        cat2label = {k: i for i, k in enumerate(self.CLASSES)}
        # load image list from file
        image_list = mmcv.list_from_file(self.ann_file)

        data_infos = []
        # convert annotations to middle format
        for image_id in image_list:
            filename = f'{self.img_prefix}/{image_id}.png'
            image = mmcv.imread(filename)
            height, width = image.shape[:2]

            data_info = dict(filename=f'{image_id}.png', width=width, height=height)

            # load annotations
            label_prefix = self.img_prefix.replace('image_2', 'label_2')
            lines = mmcv.list_from_file(osp.join(label_prefix, f'{image_id}.txt'))

            content = [line.strip().split(' ') for line in lines]
            bbox_names = [x[0] for x in content]
            bboxes = [[float(info) for info in x[4:8]] for x in content]

            gt_bboxes = []
            gt_labels = []
            gt_bboxes_ignore = []
            gt_labels_ignore = []

            # filter 'DontCare'
            for bbox_name, bbox in zip(bbox_names, bboxes):
                if bbox_name in cat2label:
                    gt_labels.append(cat2label[bbox_name])
                    gt_bboxes.append(bbox)
                else:
                    gt_labels_ignore.append(-1)
                    gt_bboxes_ignore.append(bbox)

            data_anno = dict(
                bboxes=np.array(gt_bboxes, dtype=np.float32).reshape(-1, 4),
                labels=np.array(gt_labels, dtype=np.long),
                bboxes_ignore=np.array(gt_bboxes_ignore,
                                       dtype=np.float32).reshape(-1, 4),
                labels_ignore=np.array(gt_labels_ignore, dtype=np.long))

            data_info.update(ann=data_anno)
            data_infos.append(data_info)

        return data_infos
```

```python
import mmcv
import matplotlib.pyplot as plt

def show_result(img, result, class_names, score_thr=0.3):
    img = mmcv.imshow(img)
    bbox_result, segm_result = result

    # Dibujar las predicciones de detección de objetos
    bboxes = np.vstack(bbox_result)
    labels = [
        np.full(bbox.shape[0], i, dtype=np.int32)
        for i, bbox in enumerate(bbox_result)
    ]
    labels = np.concatenate(labels)
    inds = np.where(bboxes[:, -1] > score_thr)[0]
    for i in inds:
        bbox = bboxes[i, :4]
        label = labels[i]
        color = plt.cm.get_cmap('hsv', len(class_names) + 1)(label / len(class_names))
        color = tuple([int(c * 255) for c in color[:3]])
        mmcv.imshow_det_bboxes(
            img,
            np.array([bbox]),
            np.array([label]),
            np.array([bboxes[i, -1]]),
            class_names=class_names,
            show=False,
            colors=[color],
            thickness=2,
            font_scale=0.8)

    # Dibujar las predicciones de segmentación de objetos
    if segm_result is not None:
        segms = mmcv.concat_list(segm_result)
        inds = np.where(bboxes[:, -1] > score_thr)[0]
        for i in inds:
            color_mask = np.random.randint(0, 256, (1, 3), dtype=np.uint8)
            mask = segms[i]
            img = mmcv.draw_mask(img, mask, color_mask, alpha=0.5)

    # Mostrar la imagen con las predicciones
    plt.imshow(mmcv.bgr2rgb(img))
    plt.show()
```
