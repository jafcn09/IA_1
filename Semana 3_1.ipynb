# Clasificación

```python
# conjunto de datos artificial
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=1000, n_features=2,  n_informative=2, n_redundant=0)

# (procesar el conjunto de datos)
# (análisis exploratorio de datos)
```

```python
# gráfico del conjunto de datos
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], c=y, s=5)
plt.xlabel("Variable 1") # dimension 1, X[0], feature 1, característica 1
plt.ylabel("Variable 2") # dimension 2,  X[1], feature 2, característica 2
```

```python
# dividir los datos en conjuntos entrenamiento - prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
```

```python
# definir las máquinas de aprendizaje (machine learning models)
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression()
clf_lr
```

```python
# entrenar la máquina
clf_lr.fit(X_train, y_train)
```

```python
# parámetros del modelo
print(clf_lr.coef_)
print(clf_lr.intercept_)
```

```python
# realizando las predicciones
y_pred = clf_lr.predict(X_test)
y_pred
```

```python
# medir el desempeño del modelo usando una métrica
# mide el poder predictivo/generalización del modelo
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```

```python
# calculando el error de entrenamiento
y_pred_train = clf_lr.predict(X_train)
accuracy_score(y_pred_train, y_train)
```

# clasificación usando varios modelos:
* regresión logística
* los K vecinos mas cercanos (KNN)
* Máquinas de vectores de soporte (SVM)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

clf_lr = LogisticRegression() # modelo paramétrico
clf_knn = KNeighborsClassifier() # modelo no paramétrico
clf_svm = SVC() # modelo semi-paramétrico
```

```python
# entrenar los modelos
for clf in [clf_lr, clf_knn, clf_svm]:
  clf.fit(X_train, y_train)
```

```python
# realizr las predicciones
# medir la generalización o poder predictivo de cada uno de los modelos
from sklearn.metrics import accuracy_score

for clf in [clf_lr, clf_knn, clf_svm]:

  # predicciones
  y_pred = clf.predict(X_test)

  # usando la métrica
  print(accuracy_score(y_pred, y_test))
```

# Regresión

```python
# conjunto de datos
from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 1000, n_features =1, n_informative = 1, noise=10)
```

```python
# gráfico del conjunto de datos
plt.scatter(X, y, s=0.5)
plt.xlabel("Variable 1")# X, predictor, dimension, caracteristica, feature
plt.ylabel("label") # y, target, variable objetivo, label
```

```python
#dividir los datos en conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)
```

```python
# definición de las máquinas
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

reg_lr = LinearRegression() # paramétrico
reg_knn = KNeighborsRegressor() # no paramétrico
reg_svm = SVR() # semi-paramétrico
```

```python
# entrenamiento
for model in [reg_lr, reg_knn, reg_svm]:
  model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

#realizar predicciones
for model in [reg_lr, reg_knn, reg_svm]:
  y_pred = model.predict(X_test)

  # evaluar la perfomance usando la metrica
  print(mean_squared_error(y_pred, y_test))
```

```python
# lo mismo pero en una sola linea de código
resultados = []
for model in [reg_lr, reg_knn, reg_svm]:
  resultados.append(mean_squared_error(model.fit(X_train, y_train).predict(X_test), y_test))

print(resultados)
```

```python
#plot
y_pred = reg_lr.predict(X_test)

plt.scatter(X_test, y_test, color = 'black', s=0.5)
plt.scatter(X_test, y_pred, color = 'blue', s=0.5)
```
